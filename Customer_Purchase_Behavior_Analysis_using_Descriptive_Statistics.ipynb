```python
# Complain column for Bernoulli distribution for Complain column
from scipy.stats import bernoulli

p_complain = df['Complain'].mean()
x_bernoulli = [0, 1]
pmf_bernoulli = bernoulli.pmf(x_bernoulli, p_complain)

variance = np.var(df['Complain'])

print("Probability of 0:", pmf_bernoulli[0])
print("Probability of 1:", pmf_bernoulli[1])
print("Variance:", variance)
```

```python
# Calculate mean and standard deviation for the normal distribution
mean_income_normal = df_no_outliers['Income'].mean()
std_dev_income_normal = df_no_outliers['Income'].std()

lower_bound_normal = mean_income_normal - std_dev_income_normal
upper_bound_normal = mean_income_normal + std_dev_income_normal

probability_within_range_normal = norm.cdf(upper_bound_normal, loc=mean_income_normal, scale=std_dev_income_normal) - norm.cdf(lower_bound_normal, loc=mean_income_normal, scale=std_dev_income_normal)

# Calculate expected value (mean) for the normal distribution
expected_value_income_normal = mean_income_normal


# Calculate probability of a customer having income within 2 standard deviations from the mean for normal distribution
lower_bound_normal = mean_income_normal - 2 * std_dev_income_normal
upper_bound_normal = mean_income_normal + 2 * std_dev_income_normal

probability_within_2std_normal = norm.cdf(upper_bound_normal, loc=mean_income_normal, scale=std_dev_income_normal) - norm.cdf(lower_bound_normal, loc=mean_income_normal, scale=std_dev_income_normal)

# Print results
print(f"Probability of a customer having income within 1 standard deviation from the mean (Normal): {probability_within_range_normal:.2%}")
print(f"Expected value (mean) for the normal distribution: {expected_value_income_normal:.2f}")

print(f"Probability of a customer having income within 2 standard deviations from the mean (Normal): {probability_within_2std_normal:.2%}")
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**

  

```python
from scipy.stats import zscore


# Select relevant numerical columns for correlation analysis
numerical_columns = ['Age', 'Kidhome', 'Teenhome', 'Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits',
                     'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases',
                     'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Recency']

# Remove outliers using z-score
z_scores = zscore(df[numerical_columns])
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df = df[filtered_entries]

# Calculate correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Visualize correlations using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()

```
